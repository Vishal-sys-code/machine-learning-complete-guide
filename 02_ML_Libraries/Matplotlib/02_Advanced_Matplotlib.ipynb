```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# **Colored Scatterplots**

```python
iris = pd.read_csv('/content/iris.csv')
iris.head()
```

```python
# Encoding the Species
iris['Species'] = iris['Species'].replace({'Iris-setosa': 0 ,'Iris-versicolor': 1, 'Iris-virginica': 2})
iris.head(5)
```

## **`cmap` and `alpha`**

```python
plt.figure(figsize = (10, 6))
plt.scatter(x = iris['SepalLengthCm'], y = iris['PetalLengthCm'], c = iris['Species'], cmap = 'winter', alpha = 0.7)
plt.title("Iris Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.colorbar()
plt.show()
```

## **plot size**

```python
plt.figure(figsize=(10,10))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

# **Annotations**

```python
batters = pd.read_csv('batter.csv')
print("Shape of the batters dataset: ", batters.shape)
```

```python
sample_df = batters.head(100).sample(25,random_state=5)
sample_df.sample(5)
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
plt.show()
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.scatter(x,y)
plt.text(1,5,'Point 1')
plt.text(2,6,'Point 2')
plt.text(3,7,'Point 3')
plt.text(4,8,'Point 4')
plt.show()
```

```python
plt.figure(figsize = (16,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'], s = sample_df['runs'])

for i in range(len(sample_df)):
  plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

# **Plotting Horizontal and Vertical Lines**

```python
plt.figure(figsize = (16,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'], s = sample_df['runs'])
plt.axhline(130, color = 'red')
plt.axhline(140, color = 'green')
plt.axvline(30, color = 'red')
for i in range(len(sample_df)):
  plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

# **SubPlots**

```python
batters.head()
```

```python
fig,ax = plt.subplots(figsize = (15, 6))
ax.scatter(batters['avg'], batters['strike_rate'])
ax.set_title('Scatter Plot')
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')
fig.show()
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 1, sharex = True, figsize = (10,6))
ax[0].scatter(batters['avg'], batters['strike_rate'])
ax[1].scatter(batters['avg'], batters['runs'])
ax[0].set_title("Average vs Strike Rate")
ax[0].set_ylabel("Strike Rate")
ax[1].set_title("Average vs Runs")
ax[1].set_xlabel("Runs")
ax[1].set_ylabel("Average")
fig.show()
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,6))
ax[0,0].scatter(batters['avg'], batters['strike_rate'])
ax[0,1].scatter(batters['avg'], batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['avg'])
fig.show()
```

```python
# Grid should be written carefully
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batters['avg'], batters['strike_rate'], color = 'red')
ax2 = fig.add_subplot(2,2,2)
ax2.hist(batters['runs'])
ax3 = fig.add_subplot(2,2,3)
ax3.hist(batters['avg'])
fig.show()
```

# **3D Scatter Plots**

```python
fig = plt.figure(figsize = (10,6))
ax = plt.subplot(projection = '3d')
ax.scatter3D(batters['runs'], batters['avg'], batters['strike_rate'])
ax.set_title("IPL Batsman Analysis")
ax.set_xlabel('Runs')
ax.set_ylabel('Average')
ax.set_zlabel('Strike Rate')
fig.show()
```

# **3D Surface Plots**

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
xx, yy = np.meshgrid(x, y)
print("Shape of XX: ", xx.shape)
print("Shape of YY: ", yy.shape)
```

```python
z = xx**2 + yy**2
print("Shape of z: ", z.shape)
```

```python
fig = plt.figure(figsize = (12,8))
ax = plt.subplot(projection = '3d')
cb = ax.plot_surface(xx, yy, z, cmap = 'viridis')
fig.colorbar(cb)
ax.set_title("Surface Plot")
fig.show()
```

```python
z = np.sin(xx) + np.cos(yy)
print("Shape of z: ", z.shape)
```

```python
fig = plt.figure(figsize = (12,8))
ax = plt.subplot(projection = '3d')
cb = ax.plot_surface(xx, yy, z, cmap = 'viridis')
fig.colorbar(cb)
```

# **Contour Plot**

```python
fig = plt.figure(figsize = (12,8))
ax = plt.subplot()
cb = ax.contour(xx, yy, z, cmap = 'viridis')
fig.colorbar(cb)
fig.show()
```

```python
fig = plt.figure(figsize = (12,8))
ax = plt.subplot()
cb = ax.contourf(xx, yy, z, cmap = 'viridis')
fig.colorbar(cb)
fig.show()
```
