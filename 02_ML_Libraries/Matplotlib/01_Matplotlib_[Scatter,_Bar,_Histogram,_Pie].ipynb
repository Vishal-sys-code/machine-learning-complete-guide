## 1.4 **From a pandas dataframe**


```python
batsman = pd.read_csv('/content/sharma-kohli.csv')
batsman.head(10)
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
```

## 1.5 **Plotting multiple plots**


```python
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

## 1.6 **Labels title**

```python
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

## 1.7 **Colors(hex) and Line(width and style) and marker(size)**

```python
plt.plot(batsman['index'], batsman['V Kohli'], color = 'green')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black')
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color = 'green' , linestyle = 'dashed')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dashed')
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color = 'green' , linestyle = 'dotted')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dotted')
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color = 'green' , linestyle = 'dashdot')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dashdot')
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color = 'green' , linestyle = 'dashdot', marker = 'D', markersize = 7)
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dashdot', marker = 'o', markersize = 5)
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

## 1.8 **legend -- location**

```python
plt.plot(batsman['index'], batsman['V Kohli'], color = 'green' , linestyle = 'dashdot', marker = 'D', markersize = 7, label = 'Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dashdot', marker = 'o', markersize = 5, label = 'Rohit')
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color = 'green' , linestyle = 'dashdot', marker = 'D', markersize = 7, label = 'Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dashdot', marker = 'o', markersize = 5, label = 'Rohit')
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
# plt.legend(loc = 'upper right') # Thi sill make graph dirty
plt.legend(loc = 'best')
```

## 1.9 **limiting -axes**

```python
price = [40000, 50000, 56000, 49000, 47000, 45000, 4500000]
year = [2015, 2016, 2017, 2018, 2019, 2020, 2021]
plt.plot(year, price)
plt.ylim(0, 75000)
plt.xlim(2017, 2019)
```

## 1.10 **`plt.grid`**

```python
plt.plot(batsman['index'], batsman['V Kohli'], color = 'green' , linestyle = 'dashdot', marker = 'D', markersize = 7, label = 'Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color = 'black', linestyle = 'dashdot', marker = 'o', markersize = 5, label = 'Rohit')
plt.title('V Kohli vs RG Sharma')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
# plt.legend(loc = 'upper right') # Thi sill make graph dirty
plt.legend(loc = 'best')
plt.grid()
```

## 1.11 **`plt.show`**

<div align="center">
  <h1><b>Chapter 3: Bar Chart</b></h1>
</div>

Bar Chart are used for:
- Bivariate and Univariate Analysis
- Numerical vs Categorical
- Use Case: Aggregate Analysis of groups

```python
df = pd.read_csv('/content/batsman_season_record.csv')
df
```

## 3.1 **Simple Bar chart**


```python
children = [10, 20, 30, 40, 50]
colors = ['red', 'blue', 'green', 'yellow', 'pink']
plt.bar(colors, children)
# You can use color, and labels here too
# This is the Vertical Bar Chart
```

## 3.2 **Horizontal Bar chart**

```python
children = [10, 20, 30, 40, 50]
colors = ['red', 'blue', 'green', 'yellow', 'pink']
plt.barh(colors, children)
```

## 3.3 **Multiple Bar charts**


```python
plt.bar(df['batsman'], df['2015'])
plt.bar(df['batsman'], df['2016'])
plt.bar(df['batsman'], df['2017'])
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width = 0.2)
plt.bar(np.arange(df.shape[0]), df['2016'], width = 0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width = 0.2)
```

## 3.4 **`xticks`**


```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width = 0.2)
plt.bar(np.arange(df.shape[0]), df['2016'], width = 0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width = 0.2)
plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.show()
```

## 3.5 **Problem -  Label overlapping**

```python
children = [10, 20, 40, 10, 30]
colors = ['red red red red red red', 'blue blue blue blue blue blue', 'green green green green green', 'yellow yellow yellow yellow yellow', 'pink pink pink pink pink']
plt.bar(colors, children)
# To remove the overlapping of texts
plt.xticks(rotation = 'vertical')
plt.show()
```

## 3.6 **Stacked Bar chart**

```python
# Make sure that your data is sorted or indexed properly
plt.bar(df['batsman'], df['2015'], label='2015')
plt.bar(df['batsman'], df['2016'], bottom=df['2015'], label='2016')
plt.bar(df['batsman'], df['2017'], bottom=df['2015'] + df['2016'], label='2017')

# Adding labels and title
plt.xlabel('Batsman')
plt.ylabel('Score')
plt.title('Batsman Scores Over Years (2015-2017)')

# Optional: Rotate the x-axis labels if there are many batsmen
# plt.xticks(rotation=90)

# Adding a legend
plt.legend()

# Display the plot
plt.tight_layout()  # Adjust the layout to avoid overlap
plt.show()
```

<div align="center">
  <h1><b>Chapter 5: Pie Chart</b></h1>
</div>

Pie Chart is used for:
- Univariate and Bivariate Analysis
- Categorial vs Numerical
- Use case: To find contribution on a standard scale

## 5.1 **Pie Chart on - simple data**

```python
marks = [75, 67, 40, 90, 75, 83, 82]
subjects = ['Algorithms', 'Machine Learning', 'English', 'Neural Networks', 'Database', 'Mathematics', 'Economics']
plt.pie(marks, labels = subjects)
plt.show()
```

## 5.2 **Pie Chart on - dataset**


```python
df = pd.read_csv('/content/gayle-175.csv')
df.head()
```

```python
plt.pie(df['batsman_runs'], labels = df['batsman'])
plt.show()
```

## 5.3 **Percentage and colors**

```python
plt.pie(df['batsman_runs'], labels = df['batsman'], autopct = '%0.1f%%')
plt.show()
```

```python
plt.pie(df['batsman_runs'], labels = df['batsman'], autopct = '%0.1f%%', colors = ['blue', 'green', 'yellow', 'red', 'cyan', 'orange'])
plt.show()
```

## 5.4 **`explode` and `shadow`**

```python
# explode
plt.pie(df['batsman_runs'], labels = df['batsman'], autopct = '%0.1f%%', explode = [0.3, 0, 0, 0, 0.2, 0.1])
plt.show()
```

```python
# shadow
plt.pie(df['batsman_runs'], labels = df['batsman'], autopct = '%0.1f%%', shadow = True)
plt.show()
```
