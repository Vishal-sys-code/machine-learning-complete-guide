A dataframe is a data structure constructed with rows and columns, similar to a database or Excel spreadsheet. It consists of a dictionary of lists in which the list each have their own identifiers or keys.

```python
ipl.head(2)
```

```python
# Find all the final winners
new_df = ipl[ipl['MatchNumber'] == 'Final']
new_df[['Season', 'WinningTeam']]
```

```python
# How many super over finishes have occured
new_df = ipl[ipl['SuperOver'] == 'Y'][['Season', 'SuperOver' , 'WinningTeam']]
print("Total Number of SuperOver: ", new_df.shape[0])
new_df
```

```python
# Toss winner is match winner in percentage
new_df = ipl[ipl['TossWinner'] == ipl['WinningTeam']][['TossWinner', 'WinningTeam']]
print("Toss Winner is Match Winner (in %): ", (new_df.shape[0]/ipl.shape[0])*100)
```

```python
# Movies with rating higher than 8 and votes>10000
new_df = movies[(movies['imdb_rating'] > 8) & (movies['imdb_votes'] > 10000)][['imdb_rating', 'imdb_votes']]
new_df.head(5)
```

```python
# Action movies with rating higher than 7.5
new_df = movies[(movies['genres'].str.contains('Action')) & (movies['imdb_rating'] > 7.5)][['genres', 'imdb_rating']]
print("Shape of the new dataframe is: ", new_df.shape[0])
new_df.head(5)
```

```python
# We will make a new column -> lead actor, and the lead actor will be the first actor of the actor.
# Step - 1. To remove the null values
print("Missing Values in actors: ", movies['actors'].isnull().sum())
movies.dropna(inplace = True)
```

```python
# astype -> Changing the DataType
ipl['ID'] = ipl['ID'].astype('int32') # Permanent Change
ipl.info()
```

```python
ipl['Season'] = ipl['Season'].astype('category') # Permanent Change
ipl.info()
```

```python
ipl['Margin'] = ipl['Margin'].astype('float16') # Permanent Change
ipl.info()
```
