## **1.1 What is Pandas**
Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool, built on top of the Python programming language.

[Documentation](https://pandas.pydata.org/about/index.html)

**What is Pandas Series?**

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

Screenshot (100).png

```python
import numpy as np
import pandas as pd
```

## **1.2 Series from lists**

We will get: `index` and `value`.

In pandas, string means object.

```python
# String
country = ["India", "USA", "UK", "Australia", "Russia"]
pd.Series(country)
```

```python
# Integers
sachin_tendulkar_runs = [100, 200, 131, 99, 80, 13, 45]
sachin_tendulkar = pd.Series(sachin_tendulkar_runs)
print(sachin_tendulkar)
```

Here, we got the index automatically whereas we can generate series with our own indexes. Now we will go through the custom index.

```python
# Custom Index
marks = [67, 57, 89, 100]
subjects = ['Mathematics', 'English', 'Science', 'Hindi']
"""Here, we need to show: subjects as an index and marks as the value."""
pd.Series(marks, index = subjects)
```

We can also name our series so that we can call it in further in future.

```python
# Setting a name
marks_of_cr7 = pd.Series(marks, index = subjects, name = "Marks of Cristiano Ronaldo")
print(marks_of_cr7)
```

## **1.3 Series from dict**

```python
# We can also create using the dictionary
marks = {
    'Mathematics': 67,
    'English': 57,
    'Science': 89,
    'Hindi': 100
}
marks_series = pd.Series(marks, name = "Marks of CR7")
print(marks_series)
```

```python
# index
print("Index of marks_series: ", marks_series.index)
```

## **1.5 Series using `read_csv`**

```python
# With one column
subs_dataset = pd.read_csv('/content/subs.csv')
print("Datatype of subs_dataset: ", type(subs_dataset)) # datatype: Dataframe
# Converting to the pandas Series
subs_dataset_conversion = pd.read_csv('/content/subs.csv').squeeze("columns")
print("Datatype of subs_dataset: ", type(subs_dataset_conversion)) # datatype: pandas Series
print()
subs = subs_dataset_conversion
print("--------------- Displaying the subs data ---------------")
print(subs)
```

```python
# With two columns
kohli_ipl_runs = pd.read_csv('/content/kohli_ipl.csv', index_col = 'match_no').squeeze("columns")
print("Datatype of the kohli_ipl_runs: ", type(kohli_ipl_runs))
print()
vk = kohli_ipl_runs
print("---------------- Displaying the vk: ----------------")
print(vk)
```

```python
movies = pd.read_csv('/content/bollywood.csv', index_col = 'movie').squeeze("columns")
print("---------------- Displaying the movies ----------------")
print(movies)
```

## **2.1 `series.head()`/`series.tail()`**

```python
print("Head of subs dataset: ")
print(subs.head())
print()
print("Head of the vk dataset: ")
print(vk.head(5))
print()
print("Tail of the vk dataset: ")
print(vk.tail(5))
```

## **2.2 `series.sample()`**

This is used when there is a bias in the dataset

```python
print("Sample of movies dataset: ")
print(movies.sample())
print()
print(movies.sample(5))
```

## **2.3 `value_counts()`**

It return the frequency of the value column.

```python
print(movies.value_counts())
print()
print(vk.value_counts())
```

## **2.4 `sort_values()`**

```python
# Method Chaining
print(vk.sort_values()) # Ascending Order
print()
print(vk.sort_values(ascending = False)) # Descending Order
print()
print(vk.sort_values(ascending = False).head(1)) # Top Row
print()
print(vk.sort_values(ascending = False).head(1).values) # Numpy array of runs
print()
print(vk.sort_values(ascending = False).head(1).values[0]) # Getting number from array as highest runs
```

```python
# If we want permanent changes in the series dataset then use the implace
# print(vk.sort_values(ascending = False, inplace = True))
# This will change permanently sorting to your dataset.
# Use carefully bitch
```

## **2.5 `sort_index()`**

It will sort the index, same as the `sort_values` has done the sorting to the values.

```python
movies.sort_index()
```

## **3.1 `count()`**

```python
print("Count in subs dataset: ", subs.count())
print("Count in vk dataset: ", vk.count())
print("Count in movies dataset: ", movies.count())
```

## **3.2 `series.sum`/`product`**

```python
print("Sum of the subs dataset: ", subs.sum())
print("Sum of the vk dataset: ", vk.sum())
print("Product of the subs dataset: ", subs.product())
print("Product of the vk dataset: ", vk.product())
```

## **3.3 `mean`/`median`/`mode`**

```python
print("Mean of the subs dataset: ", subs.mean())
print("Median of the subs dataset: ", subs.median())
print("Mode of the subs dataset: ", subs.mode())
print()
print("Mean of the vk dataset: ", vk.mean())
print("Median of the vk dataset: ", vk.median())
print("Mode of the vk dataset: ", vk.mode())
print()
print("Mode of the movies dataset: ", movies.mode())
print()
print("Standard Deviation of subs dataset: ", subs.std())
print("Variance of subs dataset: ", subs.var())
print()
print("Standard Deviation of vk dataset: ", vk.std())
print("Variance of vk dataset: ", vk.var())
```

## **3.4 `min`/`max`**

```python
print("Maximum Value of subs dataset: ", subs.max())
print("Minimum Value of subs dataset: ", subs.min())
print("Maximum Value of vk dataset: ", vk.max())
print("Maximum Value of vk dataset: ", subs.min())
```

## **3.5 `describe`**

```python
print("Summary of subs dataset: ")
print(subs.describe())
print()
print("Summary of vk dataset: ")
print(vk.describe())
print()
```

## **3.6 Series Indexing**

```python
# Integer Indexing
# Series work in the Positive indexing not in the negative index
x = pd.Series([12, 13, 14, 35, 46, 57, 58, 79, 9])
print("First Element: ", x[0])
print("Second Element: ", x[1])
print()
print("First Element (Movies Dataset): ", movies[0])
print()
```

```python
# Negative Indexing
print("Last Element (Movies Dataset): ", movies[-1])
print()
print("Last Element: ", x[-1])
print("Second Last Element: ", x[-2])
```

```python
# Slicing
print("Slicing the vk dataset: ")
print(vk[5:16])
```

```python
# Negative Slicing
print("Negative Slicing the vk dataset: ")
print(vk[-5:])
print()
print("Negative Slicing the movies dataset: ")
print(movies[-5:])
print()
print("Negative Slicing the movies dataset: ")
print(movies[::2])
```

```python
# Fancy Indexing
print("Fancy Indexing to vk dataset: ")
print(vk[[1,3,4,5]])
```

```python
# Indexing with the labels
print("Indexing with the labels : movies dataset ")
print(movies['Evening Shadows'])
```

## **3.7 Editing Series**

```python
# Using Indexing
print("Marks Series (Before Editing): ")
print(marks_series)
print()
marks_series[1] = 99
print("Marks Series (After Editing): ")
print(marks_series)
```

```python
# What if an index does not exist
marks_series['Physics'] = 100
print("Adding new subkect => Physics: ")
print(marks_series)
print()
marks_series['Chemistry'] = 90
print("Adding new subkect => Chemistry: ")
print(marks_series)
```

```python
# Slicing
print(" --------- Runs of Sachin Ramesh Tendulkar: --------- ")
print(sachin_tendulkar)
print()
print("--------- Runs of Sachin Ramesh Tendulkar (After Changes) : ---------")
sachin_tendulkar[4:6] = [143, 241]
print(sachin_tendulkar)
```

```python
# Fancy Indexing
sachin_tendulkar[[3,6]] = [198, 111]
print("--------- Runs of Sachin Ramesh Tendulkar (After Changes) : ---------")
print(sachin_tendulkar)
```

```python
# Using Index Label
print("Editing with the labels : movies dataset")
movies['Evening Shadows'] = 'Gigi Hadid'
print(movies)
```

```python
# Relational Operators
print(vk >= 50)
print()
print(sachin_tendulkar >= 150)
```

## **4.1 Boolean Indexing on Series**

```python
# Find no of 50's and 100's scored by Kohli
print(vk[vk >= 50].size)
```

```python
# Find Number of ducks scored by Kohli
print(vk[vk == 0].size)
```

```python
# Count number of day when I had more than 200 subs a day
print(subs[subs > 200].size)
```

```python
# Find actors who have done more than 20 movies
num_movies = movies.value_counts()
print(num_movies[num_movies > 20])
print()
print("Total Actors: ", num_movies[num_movies > 20].size)
```
