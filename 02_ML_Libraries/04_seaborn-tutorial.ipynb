# Chapter 1: Importing the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**NOTE:** Seaborn has some built-in datasets

# Chapter 2: Importing the datasets

```python
# Total Bill vs Tip dataset
tips = sns.load_dataset('tips')

# First 5 rows of the Tips dataset
tips.head()
```

# Chapter 3: Visualise the Data

```python
# Visualise the data
sns.set_theme()
sns.relplot(data = tips, x = 'total_bill', y = 'tip', col = 'time', hue = 'smoker', style = 'smoker', size = 'size')
```

# Chapter 4: Visualise the Iris dataset

```python
# loading the iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

**ScatterPlot**

```python
sns.scatterplot(x = 'sepal_length', y = 'petal_length', hue = 'species', data = iris)
```

```python
sns.scatterplot(x = 'sepal_length', y = 'petal_width', hue = 'species', data = iris)
```

# Chpater 5: Visualise the Titanic Dataset

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.shape
```

**CountPlot**

```python
sns.countplot(x = 'class', data=titanic)
```

```python
sns.countplot(x = 'survived', data = titanic)
```

**Bar Chart**

```python
sns.barplot(x='sex', y='survived', hue='class', data=titanic)
```

# Chapter 7: Correlation
* Positive Correlation
* Negative Correlation

**Heat Map**

```python
correlation = boston_df.corr()
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(correlation, cbar = True, square = True, fmt = '.1f', annot = True, annot_kws = {'size':8}, cmap = 'Blues')
```
