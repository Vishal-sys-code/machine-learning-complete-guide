# Matplotlib
Useful for making the plots

# Chapter 1: Importing library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Chapter 2: Getting datas from the numpy

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
print("The 100 values between 0 to 100: ")
print(x)
print()
print("Sine values: ")
print(y)
print()
print("Cosine values: ")
print(z)
```

# Chapter 3: Plotting the above datas and the other geometrical waves

```python
# sine wave
print("Plotting the Sine Wave: ")
plt.plot(x,y)
plt.show()
```

```python
# cosine wave
print("Plotting the Cosine Wave: ")
plt.plot(x,z)
plt.show()
```

## Labelling or Adding label to x-axis and y-axis

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title("Sine Wave")
plt.show()
```

## Parabola

```python
x = np.linspace(-10, 10, 20) # 20 values in the range of the [-10, 10]
y = x**2
plt.plot(x,y)
plt.show()
```

## Plotting with some colors and the new symbols

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show()
```

## Plotting multiple lines

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

# Chapter 4: Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'Japanese', 'Spanish', 'Russian', 'German']
people = [100, 50, 150, 40, 70]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

# Chapter 5: Pie Chart

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'Japanese', 'Spanish', 'Russian', 'German']
people = [100, 50, 150, 40, 70]
ax.pie(people, labels = languages, autopct = '%1.1f%%')
plt.show()
```

# Chapter 6: Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(x, y, color = 'g')
ax.scatter(x, z, color = 'b')
plt.show()
```

# 3d Scatter Plot

```python
fig = plt.figure()
ax = plt.axes(projection = '3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.sin(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
