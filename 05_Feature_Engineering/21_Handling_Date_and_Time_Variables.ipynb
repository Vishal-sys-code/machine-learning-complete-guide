# Working with Dates

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

## 1. Extract year

```python
date['date_year'] = date['date'].dt.year
date.sample(5)
```

## 2. Extract Month

```python
date['date_month'] = date['date'].dt.month
date.sample(5)
```

# 3. Setting Month Name

```python
date['date_month_name'] = date['date'].dt.month_name()
date.sample(5)
```

## 4. Extract Days

```python
date['date_day'] = date['date'].dt.day
date.sample(5)
```

## 5. Getting Day of Week

```python
date['date_day_of_week'] = date['date'].dt.dayofweek
date.sample(5)
```

## 6. Getting Name for the Day of Week

```python
date['date_day_of_week_name'] = date['date'].dt.day_name()
date.sample(5)
```

## 7. Dropping the data which will not be used further

```python
date.drop(columns = ['product_id', 'city_id', 'orders']).head()
```

## 8. Is Weekend?

```python
date['date_is_weekend'] = np.where(date['date_day_of_week_name'].isin(['Sunday', 'Saturday']), 1,0)
date.head()
```

## 9. Extract week of the year

```python
date['date_week'] = date['date'].dt.isocalendar().week
date.head()
```

## 10. Extract Quarter

```python
date['quarter'] = date['date'].dt.quarter
date.head()
```

## 11. Extract Semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)
date.head()
```

## 12. Extract time elapsed between dates

```python
import datetime
today = datetime.datetime.today()
today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed
np.round((today - date['date']) / np.timedelta64(1, 'D') / 30, 0)
```

# Working with Times

```python
time.info()
```

```python
time['date'] = pd.to_datetime(time['date'])
```

```python
time.info()
```

```python
time['hour'] = time['date'].dt.hour
time['min'] = time['date'].dt.minute
time['sec'] = time['date'].dt.second

time.head()
```

```python
time.drop(columns = ['msg']).head()
```

## Extract time part

```python
time['time'] = time['date'].dt.time
time.head()
```

## Time Difference

```python
today - time['date']
```

## Time Difference (in seconds, minutes and hours)

```python
print("Time Difference (in seconds): ")
print((today - time['date'])/np.timedelta64(1,'s'))
print()
print("Time Difference (in minutes): ")
print((today - time['date'])/np.timedelta64(1,'m'))
print()
print("Time Difference (in hours): ")
print((today - time['date'])/np.timedelta64(1,'h'))
```
