```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/customer.csv')
```

```python
df.sample(5)
```

```python
df = df.iloc[:,2:]
```

```python
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:2], df.iloc[:,-1], test_size = 0.2)
```

```python
X_train
```

# Ordinal Encoder

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = [['Poor', 'Average', 'Good'], ['School', 'UG', 'PG']])
```

```python
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```

```python
print("X-train: ")
print(X_train)
print("--------------------")
print("X-test: ")
print(X_test)
```

```python
oe.categories_
```

# Label Encoder
The last column of the Purchase is in the categorical form we need to convert it into the label or numerical value.

Label encoding is explicitly used when you are working with the output columns.

Encode target labels with the value range between `[0 to n_classes-1]`

This transformer should be uswed to encode the target_values i.e. `Y` and not the input `X`.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit(y_train)
```

```python
le.classes_
```

```python
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```
