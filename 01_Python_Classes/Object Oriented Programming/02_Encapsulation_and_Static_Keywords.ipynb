Q. Write OOP classes to handle the following scenarios:
- A user can create and view 2D coordinates.
- A user can find out the distance between 2 coordinates.
- A user can find the distance of a coordinate from origin.
- A user can check if a point lies on a given line.
- A user can find the distance between a given 2D point and a given line.

BrainStorming: Find that if two lines intersect each other or not.

````python
class point:
  def __init__(self, x, y):
    self.x_coordinate = x
    self.y_coordinate = y
  # Magic Method to print in the strings
  def __str__(self):
    return f"(x, y) is: <{self.x_coordinate},{self.y_coordinate}>"
  # self = (x1,y1) and other = (x2, y2)
  def euclidean_distance(self, other):
    distance = ((self.x_coordinate - other.x_coordinate)**2 + (self.y_coordinate - other.y_coordinate)**2)**0.5
    return f"The Euclidean distance between <{self.x_coordinate,self.y_coordinate}> and <{other.x_coordinate,other.y_coordinate}> are: {distance}"
  def origin_distance(self):
    distance = (self.x_coordinate**2 + self.y_coordinate**2)**0.5
    # distance = self.euclidean_distance(self,point(0,0))
    return f"The Euclidean distance between <{self.x_coordinate,self.y_coordinate}> and origin are: {distance}"
  """
  To check if the points are on the line, we need to satisfy the line equation.
  We need to substitute the values of points (x,y,z) in the line equation, if it
  comes to 0 then it lies on the line, otherwise it doesn't.
  """
  def check_on_line(self, line):
    if line.A*self.x_coordinate + line.B*self.y_coordinate + line.C == 0:
      return f"<{self.x_coordinate,self.y_coordinate}> lies on the line."
    else:
      return f"<{self.x_coordinate,self.y_coordinate}> doesn't lie on the line."
  def shortest_distance(self, line):
    numerator = abs(line.A*self.x_coordinate + line.B*self.y_coordinate + line.C)
    denominator = (line.A**2 + line.B**2)**0.5
    return f"The shortest distance is: {numerator/denominator}"

class line:
  def __init__(self, A, B, C):
    self.A = A
    self.B = B
    self.C = C
  def __str__(self):
    return f"{self.A}x + {self.B}y + {self.C} = 0"
````

### Multiple Reference

````python
p = Person()
q = p

"""
We are not doing like p = Person(). Technically when we are just calling the Person().
p is like a variable in which it store the memory address in the object.
p contains the address of the object.
p is the reference variable that contains the memory or the address of the object.

Person() itself will also work

p = Person()
q = p

Here, q and p both are pointing to the same class Person().
"""

print("Id of p: ", id(p))
print("Id of q: ", id(q))
````

### Change attribute value with the help of 2nd Object

````python
print("Name (p): ", p.name)
print("Name (q):", q.name)
q.name = 'Dahlia'
print("Name (p): ", p.name)
print("Name (q):", q.name)
````

**Explanation:** Initially, both `p.name` and `q.name` are `'Vayishu'` because both `p` and `q` refer to the same object. When you modify `q.name = 'Dahlia'`, you're changing the `name` attribute of the object both `p` and `q` point to. Since they reference the same object, the change reflects in both `p` and `q`. After the modification, both `p.name` and `q.name` will show `'Dahlia'`. This demonstrates that in Python, when two variables reference the same object, changes made through one variable affect the other, as they point to the same memory.

````python
class Person:
  def __init__(self, name, gender):
    self.name = name
    self.gender = gender

# Outside the class, and it is a function
def greet(person):
  print(f"Hello, my name is: {person.name}, and my gender is: {person.gender}")

p = Person('Vayishu', 'male')
greet(p)
````

**Explanation:** In the code, the `Person` class is defined with an `__init__` method to initialize the `name` and `gender` attributes when an object is created. The `greet` function, which is outside the class, accepts a `person` object as a parameter. It prints a greeting using the `name` and `gender` attributes of the passed `person` object. An instance of the `Person` class is created with the name "Vayishu" and gender "male" and is assigned to variable `p`. Finally, the `greet(p)` function call prints the greeting for the `p` object.

````python
class Person:
  def __init__(self, name, gender):
    self.name = name
    self.gender = gender

# Outside the class, and it is a function
def greet(person):
  print(id(person))
  print(f"Hello, my name is: {person.name}, and my gender is: {person.gender}")

p = Person('Vayishu', 'male')
print(id(p))
greet(p)

"""
We have sent a copy of the object.
This means that we have sent the address of the object as a reference.
Thats the reason the id are same.
"""
````

**Explanation:**
This code defines a `Person` class with two attributes: `name` and `gender`. When a `Person` object is created, these attributes are set using the `__init__` method. The `greet` function accepts a `Person` object as an argument and prints its unique memory ID and a greeting message using the object's `name` and `gender`.

- `p = Person('Vayishu', 'male')` creates a `Person` object.
- `print(id(p))` shows the unique memory ID of the object `p`.
- `greet(p)` calls the function and prints the details of `p`.

This demonstrates object creation and function interaction in Python.

````python
class Person:
  def __init__(self, name, gender):
    self.name = name
    self.gender = gender

# Outside the class, and it is a function
def greet(person):
  print(id(person))
  print(f"Hello, my name is: {person.name}, and my gender is: {person.gender}")
  person.name = 'Dahlia'
  print(person.name)

p = Person('Vayishu', 'male')
print(id(p))
greet(p)
print(p.name)
````

**Explanation:**
This code defines a `Person` class with `name` and `gender` attributes. The `greet` function takes a `Person` object, prints its memory ID, and greets using the `name` and `gender`. It then changes the `name` of the person to "Dahlia" and prints the updated name.

- `p = Person('Vayishu', 'male')` creates a `Person` object `p`.
- `print(id(p))` shows the memory ID of `p`.
- `greet(p)` prints the details of `p` and changes its `name`.
- Finally, `print(p.name)` shows the updated `name` ("Dahlia").

This demonstrates object manipulation and function interaction.

### Mutability of an Object


````python
class Person:
  def __init__(self, name, gender):
    self.name = name
    self.gender = gender

# Outside the class, and it is a function
def greet(person):
  person.name = 'Dahlia'
  return person

p = Person('Vayishu', 'male')
print(id(p))
p1 = greet(p)
print(id(p1))

"""
Here, the memory address depends on the mutability of the object.
If the object is mutable then address will be same, else it will be different.
"""
````

**Explanation:**
This code defines a `Person` class with attributes `name` and `gender`. The `greet` function modifies the `name` attribute of the given `Person` object to "Dahlia" and then returns the updated object.

- `p = Person('Vayishu', 'male')` creates a `Person` object `p`.
- `print(id(p))` prints the memory ID of `p`.
- `p1 = greet(p)` calls the `greet` function, which changes `p`'s `name` to "Dahlia" and returns the modified object as `p1`.
- `print(id(p1))` prints the memory ID of `p1`, which is the same as `p` because they refer to the same object.

This shows that objects are passed by reference in Python.

Now, here we need a special variable to introduce to this worls that is the static variable.

* `Instance Variable:` Variable of Objects

* `Static Variable:` Variable of Classes

The major difference between them is:
* In instance variable, all variables will be the different for each object.
* In static variable, all variables will be the same for each object.


You are making a Banking Machine. Here are some tips:
* **Name of the customer**: Instance Variable.
* **IFSC Code**: Static variable.
* **Balance**: Instance Variable.

You are making a College Website. Here are some tips:
* **College Name**: Static Variable.
* **Student of CGPA**: Instance Variable.
* **University Name**: Instance Variable.
* **Name of the Student:** Static Variable.


**NOTES:**

* **Static variable will be outside the constructor.**
* **We use class name to initialise it, where as for instance variable we use object name to initialise it.**
