It is a relationship where one class owns the other class.

`Customer HAS-A Address`: In this, the customer class has-a relationship with the address. Where we can say that the customer class is owning the address class.



````python
# Customer Class
class Customer:
  def __init__(self, name, gender, address):
    self.name = name
    self.gender = gender
    self.address = address
  """
  Here, the address is the complex entity, It contains: Home Number, Street Address,
  City, State, Zip Code.
  So for this we will make the seperate class for the Address.
  """
  def print_customer_details(self):
    print("Name of the Customer: ", self.name)
    print("Gender: ", self.gender)
  def print_address(self):
    print("Street Number: ", self.address.street_number)
    print("Street Name: ", self.address.street_name)
    print("City: ", self.address.city)
    print("State: ", self.address.state)
    print("Zip Code: ", self.address.zip_code)

# Address Class
class Address:
  def __init__(self, street_number, street_name, city, state, zip_code):
    self.street_number = street_number
    self.street_name = street_name
    self.city = city
    self.state = state
    self.zip_code = zip_code


"""
We will create a customer class object. We will provide: Name, Gender, Address.
However first we have to make the address object then we will pass that object
into the customer object.
"""
# THIS SCENARIO IS CALLED AGGREGATION
address1 = Address("123", "Main Street", "New York", "NY", "10001")
customer1 = Customer("Gabby Pham", "Female", address1)
customer1.print_customer_details()
print()
customer1.print_address()
````

# Method Overriding

Method Overriding is when a child class provides its own version of a method that already exist in its parent class. It allows the child class to change or extend the behavior of the parent class's method.

It means in short: If a parent and child classes have the same method, the biggest question will be which one will be called: Method from the Child Class only. This is called the Method Overriding.

For example, the child class can define a method with the same name as in the parent class but with different behavior.

Same method, Only the method of child class is being executed.

````python
class Phone:
  def __init__(self, price, brand, camera):
    print("Inside Phone Constructor: Parent Class")
    self.__price = price # Private Variable
    self.brand = brand
    self.camera = camera
  # Getter Method
  def buy(self):
    print("Buying a Phone")

class SmartPhone(Phone):
  def buy(self):
    print("Buying a SmartPhone")

s = SmartPhone(90000, "Apple", 15)
s.buy()
````

# `super` Keyword

The `super` keyword in python is used to call a method or constructor of the parent class from the child class.

This is commonly used for:
- Access Parent Class Method
- Initialize Parent Class

````python
class Phone:
  def __init__(self, price, brand, camera):
    print("Inside Phone Constructor: Parent Class")
    self.__price = price
    self.brand = brand
    self.camera = camera
  def buy(self):
    print("Buying a Phone: Method of Parent Class")

class SmartPhone(Phone):
  def buy(self):
    print("Buying a SmartPhone: Method of Child Class")
    # Syntax to call the method of parent class
    super().buy()

s = SmartPhone(90000, "Apple", 15)
s.buy()
````

### Super -> Constructor

````python
# Super -> Constructor
class Phone:
  def __init__(self, price, brand, camera):
    print("Inside Phone Constructor: Parent Class")
    self.__price = price
    self.brand = brand
    self.camera = camera

class SmartPhone(Phone):
  def __init__(self, price, brand, camera, os, ram):
    super().__init__(price, brand, camera) # Calling Parent Constructor
    self.os = os
    self.ram = ram
    print("Inside SmartPhone Constructor: Child Class")

s = SmartPhone(90000, "Apple", 15, "iOS", 8)
print("Operating System: ", s.os)
print("Brand: ", s.brand)
````

This code demonstrates **constructor chaining** using the **`super`** keyword, where the child class `SmartPhone` inherits from the parent class `Phone`. The parent class initializes attributes like `price`, `brand`, and `camera`, while the child class adds additional attributes (`os` and `ram`). The **`super().__init__()`** call in the child class ensures that the parent class constructor is executed first, initializing its attributes before the child class handles its own. When an object of `SmartPhone` is created, both constructors execute in sequence, and the object's attributes are printed, showcasing inheritance and reusability of code.

### Using Super Outside the class

````python
# Using Super Outside the class
class Phone:
  def __init__(self, price, brand, camera):
    print("Inside Phone Constructor: Parent Class")
    self.__price = price
    self.brand = brand
    self.camera = camera
  def buy(self):
    print("Buying a Phone: Method of Parent Class")

class SmartPhone(Phone):
  def buy(self):
    print("Buying a SmartPhone: Method of Child Class")

s = SmartPhone(90000, "Apple", 15)
s.super().buy()
````

Super can only be written in the child class.

## Single Inheritance

Single Inheritance is a type of inheritance in object oriented programming where a child class (subclass) inherits properties and methods from one parent class (superclass).

````python
class Phone:
  def __init__(self, price, brand, camera):
    print("Inside Phone Constructor: Parent Class")
    self.__price = price
    self.brand = brand
    self.camera = camera
  def buy(self):
    print("Buying a Phone")

class SmartPhone(Phone):
  pass

SmartPhone(100000, "Apple", "15px").buy()
````

### Multi Level Inheritance
Multilevel inheritance is a type of inheritance in OOP where a class inherits from another class, and then another class inherits from it, forming a chain of inheritance.

````python
class Product:
  def review(self):
    print("Product Customer Review: Method of the Grandfather Class")

class Phone(Product):
  def __init__(self, price, brand, camera):
    print("Inside Phone Constructor: Constructor of the Parent Class")
    self.__price = price
    self.brand = brand
    self.camera = camera
  def buy(self):
    print("Buying a Phone: Method of the Parent Class")

class SmartPhone(Phone):
  pass

s = SmartPhone(90000, "Apple", "15px")
s.buy()
s.review()
````

### Hierarchial Inheritance
Hierarchial inheritance is a type of inheritance in OOP where multiple child classes inherit from a single parent class.

This means that a single parent class serves a a base, and multiple subclasses shares its properties and methods.

````python
class Phone(Product):
  def __init__(self, price, brand, camera):
    print("Inside Phone Constructor: Constructor of the Parent Class")
    self.__price = price
    self.brand = brand
    self.camera = camera
  def buy(self):
    print("Buying a Phone: Method of the Parent Class")

class SmartPhone(Phone):
  pass

class FeaturePhone(Phone):
  pass

sp = SmartPhone(100000, "Apple", "15px").buy()
fp = FeaturePhone(70000, "Samsung", "15px").buy()
````

### Multiple Inheritance
Multiple inheritance is a type if inheritance in OOP where a child class inherits from two or more parent classes.

````python
# First Parent
class Phone(Product):
  def __init__(self, price, brand, camera):
    print("Inside Phone Constructor: Constructor of the First Parent Class")
    self.__price = price
    self.brand = brand
    self.camera = camera
  def buy(self):
    print("Buying a Phone: Method of the First Parent Class")

# Second Parent
class Product:
  def review(self):
    print("Customer Review: Method of the Second Parent Class")

class SmartPhone(Phone, Product):
  pass

s = SmartPhone(20000, "Apple", '15px')
s.buy()
s.review()
````

#### The Diamond Problem

````python
# https://stackoverflow.com/questions/56361048/what-is-the-diamond-problem-in-python-and-why-its-not-appear-in-python2
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor: Constructor of the First Parent Class")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone: Method of the First Parent Class")

class Product:
    def buy(self):
        print ("Product buy method: Method of the Second Parent Class")

# Method resolution order
class SmartPhone(Phone,Product):
    pass

s=SmartPhone(20000, "Apple", 12)
s.buy()
````

The diamond problem executes the child class in the order. Which ever class will be written first will be executed first, to avoid the ambiguity. It doesnt work in the Java. This is called the Method Resolution Order.

````python
# https://stackoverflow.com/questions/56361048/what-is-the-diamond-problem-in-python-and-why-its-not-appear-in-python2
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor: Constructor of the First Parent Class")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone: Method of the First Parent Class")

class Product:
    def buy(self):
        print ("Product buy method: Method of the Second Parent Class")

# Method resolution order
class SmartPhone(Product, Phone):
    pass

s=SmartPhone(20000, "Apple", 12)
s.buy()
````

# Polymorphism

Polymorphism refers to the ability of a single function, method or operator to behave differently based on the object or data type it is acting upon. It allows different classes to share the same method name but implemented it differently, enabling code reusability and flexibility.

We will learn:
- Method Overriding [We learned above]
- Method Overloading
- Operator Overloading

## Method Overriding: Run Time Polymorphism
Method Overriding is a feature in OOP where a child class provides its own implementation of a method that is already defined in its parent class.

## Method Overloading: Compile Time Polymorphism
Method overloading is a feature in OOP where methods with the same name exist in the same class but with different parameters. This allows a method to be called with different kinds of arguments and the appropiate method is invoked based on the parameters passed.

Python doesn't support method overloading in the traditional sense as in other languages like: C++, Java.

## Difference between Method Overloading and Method Overriding
- **Method Overloading:** Same class, different parameters.
- **Method Overriding:** If subclass has the same method declared in the parent class.

## Operator Overloading
Operator overloading is a feature in object-oriented programming that allows operators (like +, -, *, ==, etc.) to be customized to work with user-defined objects. By overloading operators, you can define how the operators behave when applied to instances of a class, enabling intuitive operations on those objects.

In Python, operator overloading is achieved by defining special methods (also called magic methods or dunder methods) in your class. These methods are prefixed and suffixed with double underscores (e.g., __add__, __eq__, etc.).

Key Magic Methods for Common Operators:
- __add__(self, other): For + operator.
- __sub__(self, other): For - operator.
- __mul__(self, other): For * operator.
- __eq__(self, other): For == operator.
- __lt__(self, other): For < operator.
- __str__(self): For print() and string representation.
- __truediv__(self, other): For / operator.

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # Overloading the == operator
    def __eq__(self, other):
        return self.name == other.name and self.age == other.age

# Creating Person objects
p1 = Person("Alice", 30)
p2 = Person("Alice", 30)

# Using the overloaded == operator
print(p1 == p2)  # Outputs: True
````

````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overloading the + operator
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def __repr__(self):
        return f"Point({self.x}, {self.y})"

# Creating Point objects
p1 = Point(2, 3)
p2 = Point(4, 5)

# Using the overloaded + operator
result = p1 + p2
print(result)  # Outputs: Point(6, 8)
````
