In Python, `self` is a reference to the current instance of a class, and it is needed to access the attributes and methods of that instance. When you create an object of a class, `self` allows each object to have its own separate data (like variables) and methods. It is important because it ensures that the instance is correctly tied to its attributes and actions, allowing you to interact with and modify the object's state. You use `self` inside methods to refer to the specific object that is calling the method, and it must always be the first parameter in instance methods. Without `self`, Python wouldn’t know which object’s data to operate on.

````python
class AutomaticTellerMachine:
  # Making a constructor to declare and define variables or data members
  def __init__(self):
    self.balance = 0
    self.pin = ''
    self.menu()

  # Adding Functionalities or Functions or Behaviors
  def menu(self):
    user_input = input("""
    Hi! How can I help you?
    1. Press 1 to create PIN
    2. Press 2 to Change PIN
    3. Press 3 to check balance
    4. Press 4 to withdraw money
    5. Press 5 to Exit
    """)
    if user_input == '1':
      self.create_pin()
    elif user_input == '2':
      self.change_pin()
    elif user_input == '3':
      self.check_balance()
    elif user_input == '4':
      self.withdraw_money()
    else:
      exit()

  def create_pin(self):
    user_pin = input("Enter your PIN: ")
    self.pin = user_pin
    user_balance = int(input("Enter your Balance: "))
    self.balance = user_balance
    print("PIN Created Successfully.......")
    self.menu() # For continuity

  def change_pin(self):
    old_pin = input("Enter your old PIN: ")
    if old_pin == self.pin:
      new_pin = input("Enter your new PIN: ")
      self.pin = new_pin
      print("PIN Changed Successfully.......")
      self.menu() # For continuity
    else:
      print("Incorrect PIN")
      self.menu() # For continuity
  def check_balance(self):
    user_pin = input("Enter your PIN:")
    if user_pin == self.pin:
      print(f"Your balance is: {self.balance}")
      self.menu() # For continuity
    else:
      print(f"Your PIN is incorrect!!!, Check Again MF...")
      self.menu() # For continuity
  def withdraw_money(self):
    user_pin = input("Enter your PIN: ")
    if user_pin == self.pin:
      amount = int(input("Enter the amount you want to withdraw: "))
      if amount > self.balance:
        print("Insufficient Balance")
      else:
        self.balance = self.balance - amount
        print(f"You have sufficient balance, Your reamining balance is : {self.balance}")
        self.menu() # For continuity
    else:
      print(f"Your PIN is incorrect!!!, Check Again MF...")
      self.menu() # For continuity
````

## Creating out own datatype

#### We will make a datatype => fraction

````python
class Fraction:
  # Parameterized Constructor
  def __init__(self, numerator, denominator):
    self.numerator = numerator
    self.denominator = denominator
  # Magic Method 2
  def __str__(self):
    return f"{self.numerator}/{self.denominator}"
  # Magic Method 3 [other -> second fraction]
  def __add__(self, other):
    new_numerator = self.numerator * other.denominator + other.numerator * self.denominator
    new_denominator = self.denominator * other.denominator
    return f"{new_numerator}/{new_denominator}"
  # Magic Method 4
  def __sub__(self, other):
    new_numerator = self.numerator * other.denominator - other.numerator * self.denominator
    new_denominator = self.denominator * other.denominator
    return f"{new_numerator}/{new_denominator}"
  # Magic Method 5
  def __mul__(self, other):
    new_numerator = self.numerator * other.numerator
    new_denominator = self.denominator * other.denominator
    return f"{new_numerator}/{new_denominator}"
  # Magic Method 6
  def __truediv__(self, other):
    new_numerator = self.numerator * other.denominator
    new_denominator = self.denominator * other.numerator
    return f"{new_numerator}/{new_denominator}"
  def convertToDecimal(self):
    return (self.numerator/self.denominator)
````

````python
fraction = Fraction(1,2)
fraction1 = Fraction(3,4)
print("Fraction is: ", fraction)
print("Fraction is: ", fraction1)
````

````python
print(f"Addition of two fractions: ({fraction} + {fraction1}) = {fraction + fraction1}")
print(f"Subtraction of two fractions: ({fraction} - {fraction1}) = {fraction - fraction1}")
print(f"Multiplication of two fractions: ({fraction} * {fraction1}) = {fraction * fraction1}")
print(f"Division of two fractions: ({fraction} / {fraction1}) = {fraction / fraction1}")
print(f"Decimal Values of two fractions are: ({fraction}) = {fraction.convertToDecimal()}")
print(f"Decimal Values of two fractions are: ({fraction1}) = {fraction1.convertToDecimal()}")
````
