# Chapter 6: Manipulating a DataFrame

```python
# Removing a row
boston_dataset.drop(index = 0, axis = 0)
```

```python
# Drop a column
boston_dataset.drop(columns = 'zn', axis = 1)
```

```python
# Locating a row using the index value
boston_dataset.iloc[1]
```

```python
# Locating a particular column
print(boston_dataset.iloc[:,0]) # first coulmn
print(boston_dataset.iloc[:,1]) # second coulmn
print(boston_dataset.iloc[:,2]) # third coulmn
print(boston_dataset.iloc[:,-1]) # last coulmn
```
