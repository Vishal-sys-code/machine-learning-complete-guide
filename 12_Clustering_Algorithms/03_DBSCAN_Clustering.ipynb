# DENSITY BASED SPATIAL CLUSTERING OF APPLICATIONS WITH NOISE [DBSCAN]

```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt
X = np.array([[1,2], [2,2], [2,3], [8,7], [8,8], [25, 80]])
```

```python
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
db = DBSCAN(eps = 3, min_samples = 2)
db.fit(X)
print("Labels: ", db.labels_)
```

```python
# After increasing epsilon, that is we are increasing the radius of a circle.
db = DBSCAN(eps = 10, min_samples = 2)
db.fit(X)
print("Labels: ", db.labels_)
```

## Concentric Circles Dataset

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
import numpy as np

# Create a concentric circle dataset
X, _ = make_circles(n_samples = 500, factor = .5, noise = 0.03, random_state = 4)

# Apply DBSCAN to the dataset
dbscan = DBSCAN(eps = 0.1, min_samples = 5)
clusters = dbscan.fit_predict(X)

# Plotting
plt.scatter(x = X[:,0], y = X[:,1], c = clusters, cmap = 'viridis', marker = 'o')
plt.title("DBSCAN Clustering on Concentric Circles Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
