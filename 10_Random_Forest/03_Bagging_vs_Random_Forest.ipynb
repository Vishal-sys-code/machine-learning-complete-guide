```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.tree import plot_tree
```

```python
X,y = make_classification(n_features = 5, n_redundant = 0, n_informative = 5, n_clusters_per_class = 1)
```

```python
df = pd.DataFrame(X, columns = ['col1', 'col2', 'col3', 'col4', 'col5'])
df['target'] = y
print("Shape of the DataFrame is: ", df.shape)
df.head()
```

## Bagging Classifier: Tree based Sampling Split

```python
bag = BaggingClassifier(max_features = 2)
bag.fit(df.iloc[:,:5], df.iloc[:,-1])
plt.figure(figsize = (12, 12))
plot_tree(bag.estimators_[0])
plt.show()
```

## Random Forest: Node Level Sampling Split

```python
random_forest = RandomForestClassifier(max_features = 2)
random_forest.fit(df.iloc[:,:5], df.iloc[:,-1])
plt.figure(figsize = (12, 12))
plot_tree(random_forest.estimators_[0])
plt.show()
```

Due to this node based sampling, this creates the extra randomness, and this is why the random forest outperforms than the bagging.
