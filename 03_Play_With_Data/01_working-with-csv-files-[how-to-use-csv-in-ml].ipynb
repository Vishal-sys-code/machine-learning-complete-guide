# 14. Converters
Convertors are very useful when we need to convert something in data into different thing.
We need to define a function for converting and we provide that function in dictionary along with column name in which we want to make some chnages.

```python
pd.read_csv("../input/dataforworkingwithcsv/IPL Matches 2008-2020.csv").head(3)
```

**Here we will want to convert name of team1 and team2 into their short forms**

```python
def shortNames(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    elif name == "Kolkata Knight Riders":
        return "KKR"
    elif name == "Mumbai Indians":
        return "MI"
    elif name == "Kings XI Punjab":
        return "KXIP"
    elif name == "Chennai Super Kings":
        return "CSK"
    elif name == "Rajasthan Royals":
        return "RR"
    else:
        return name
```

```python
shortNames("Royal Challengers Bangalore")
```

```python
pd.read_csv("../input/dataforworkingwithcsv/IPL Matches 2008-2020.csv", converters = {'team1': shortNames})
```
