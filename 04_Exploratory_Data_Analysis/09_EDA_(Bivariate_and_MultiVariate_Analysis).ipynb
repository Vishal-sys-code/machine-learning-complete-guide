# Bivariate and MultiVariate Analysis
* If we do the independent analysis on the 2 columns is called the Bivariate Analysis.
* If we do the independent analysis on more than 2 columns is called the Multivariate Analysis.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

We will use 4 datasets to explore this topic
Datasets are:
* Tips Dataset [Built-in Seaborn Restaurant Data]
* Iris Dataset
* Titanic Training Data
* US Flights dataset [Built-in Seaborn Flights Dataset]

Let's import all the datasets and check them all

```python
# Importing the datasets
tips = sns.load_dataset('tips')
iris = pd.read_csv('/content/iris.csv')
titanic = pd.read_csv('/content/train.csv')
flights = sns.load_dataset('flights')
```

## 1. Scatterplot (Numerical - Numerical)

Scatterplot is used when one of the feature is numerical and other is also the numerical.

`factors = (numerical, numerical)`

We will take the dataset -> tips dataset and we will take the features: total_bill and tip as they are the numerical values and both are in USD so lets see what is the outcome.

```python
sns.scatterplot(x = tips['total_bill'], y = tips['tip'])
```

This was the bivariate analysis. Now we will do the multivariate analysis.
There is a hidden parameter called: `hue`.
We will take a more feature that is: gender or sex, so here we have now three features: `total_bill`, `tips`, `sex`. We need sex or gender because we need to check which gender has paid more.

[I am not a misogynist but yes I have crave to see this, Gender quality should be followed under the precision of the Laws of Nature.]

```python
sns.scatterplot(x = tips['total_bill'], y = tips['tip'], hue = tips['sex'])
```

This is the multivariate analysis in which the three parameters or features are in the graph plot. Fabulous!!!

Lets, suppose we need to identify the smoking customers in restaurant-tips dataset, we need to take the feature `smoker` in counter.

So, we will use the feature `smoker` to identify who are smokers and who are not.

```python
sns.scatterplot(x = tips['total_bill'], y = tips['tip'], hue = tips['sex'], style = tips['smoker'])
```

This is again a multivariate analysis in which there are four parameters or the features inside it.

Let's take one more feature that is the `size` feature which denotes what is the size of the family as a customer.
Suppose it has written 2, it means that there are two peoples in the family who came in that restaurant.

```python
sns.scatterplot(x = tips['total_bill'], y = tips['tip'], hue = tips['sex'], style = tips['smoker'], size = tips['size'])
```

This is how you can do the multivariate analysis. Now there are 5 information/parameters or the features are in this graph plot.

## 2. Barplot (Numerical - Categorial)

Barplot is used when one data is Numerical and other is the Categorial.

`factors = (Numerical, Categorial)`

We will use the titanic dataset for this.
We need to analyze with the help of the Pclass what is the average age of passengers travelling in them.

```python
sns.barplot(x = titanic['Pclass'], y = titanic['Age'])
```

The middle black rod is the Confidence Interval. We can study in statistics.

Let's see the features: fares and the pclass

```python
sns.barplot(x = titanic['Pclass'], y = titanic['Fare'])
```

Lets do the multivariate analysis.

What is the average fare according to the gender. In that case we will use hue.

```python
sns.barplot(x = titanic['Pclass'], y = titanic['Fare'], hue = titanic['Sex'])
```

```python
sns.barplot(x = titanic['Pclass'], y = titanic['Age'], hue = titanic['Sex'])
```

## 3. BoxPlot (Numerical - Categorical)

It is a 5 way summary where we get the outliers also. Lets do the boxplot between the categorical and numerical.

```python
sns.boxplot(x = titanic['Sex'], y = titanic['Age'])
```

```python
sns.boxplot(x = titanic['Sex'], y = titanic['Age'], hue = titanic['Survived'])
```

## 4. Distplot (Numerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived'] == 0]['Age']) # Dead
sns.distplot(titanic[titanic['Survived'] == 1]['Age']) # Alive
```

```python
sns.distplot(titanic[titanic['Survived'] == 0]['Age'], hist = False) # Dead
sns.distplot(titanic[titanic['Survived'] == 1]['Age'], hist = False) # Alive
```

* Blue Curve = Death
* Orange Curve = Alive

If your age is less, then the probabilty of death is lesser than the probability to be safe and alive. If you age is 25-30 yrs, the probablity of death is higher than the probability to be safe and alive.

## 5. HeatMap (Categorical - Categorical)

Heatmap is used to identify the relation between two categorial features.

```python
titanic.head(3)
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

We are getting the numbers, not we are not getting the percentage of the deaths that we need to analyze in the heatmap. So, we need to calculate the percetange and we need to use the `groupby` that is in the pandas dataframe.

```python
titanic.groupby('Pclass')['Survived'].mean()*100
```

This code gives the percentages of the people survived from each class...

Lets plot the Percentages in the bar plot.

```python
(titanic.groupby('Pclass')['Survived'].mean()*100). plot(kind = 'bar')
```

```python
titanic.groupby('Sex')['Survived'].mean()*100
```

```python
titanic.groupby('Embarked')['Survived'].mean()*100
```

```python
titanic.groupby('Embarked')['Pclass'].mean()*100
```

## 8. LinePlot (Numerical - Numerical)
It is a special case for the Scatterplot.
When the quantity of the x-axis is the time based.

Example: Flights, Stock market, Covid, etc.

```python
flights.head()
```

```python
flights.groupby('year').sum(numeric_only=True)
```

```python
flights.groupby('year').sum(numeric_only=True).reset_index()
```

```python
new = flights.groupby('year').sum(numeric_only=True).reset_index()
sns.lineplot(x = new['year'], y = new['passengers'])
```

```python
flights.pivot_table(values = 'passengers', index = 'month', columns = 'year')
```

```python
sns.heatmap(flights.pivot_table(values = 'passengers', index = 'month', columns = 'year'))
```

```python
sns.clustermap(flights.pivot_table(values = 'passengers', index = 'month', columns = 'year'))
```
